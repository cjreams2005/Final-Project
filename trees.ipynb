In this notebook, the trails in Pittsburgh neighborhoods is evaluated. I cleaned up the dataset so I have key information needed for our metric. We were aiming to find the best neighborhood to have a nice and relaxing walk when you're in that kind of mood for a reset and refresh.

```python
# load pandas
import pandas as pd
import numpy as np

# load data
tree_data = pd.read_csv("data/trees.csv", index_col="_id")

tree_data.head(5)
```

```python
## makes new dataframe of how many trees are in each nieghborhood

tally = tree_data['neighborhood'].value_counts().rename_axis('neighborhood').reset_index(name='count')

print(tally.head(10))

sorted = tally.sort_values(by='count', ascending=False)

file_path = 'final_metrics/sorted_trees_counts_by_neighborhood.csv'
sorted.to_csv(file_path, index=False)
#tally
```

```python
## display bar chart of data so its easier to see
import matplotlib.pyplot as plt

df = pd.DataFrame(tally).head(10)

df.plot(x='neighborhood', y='count', kind='bar')
plt.title('Trees in Pittsburgh Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Trees')
plt.show()
```

From the amount of trees point of view, Squirrel Hill South seems to have the most trees throughout the neighborhoods of Pittsburgh. Having a the most trees helps with cleaner air and a calmer environment. Thus, as for the trees metric, Squirrel Hill South beats out the rest of the other Pittsburgh neighborhoods. If our other metrics do not align with Squirrel Hill South, then the bar chart provides the top 10 neighborhoods with the most trees.
