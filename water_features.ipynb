# Introduction
The objective of this project is to find the best neighborhood in Pittsburgh. We decided that to find it we would use metrics related to the best place to take a walk. I decided as my metric I would see which neighborhood has the most water features. When I'm taking walks I love seeing water features. There aesthetically pleasing and drinking fountains are great for walks when it hot out. 

```python
import pandas as pd
water_features = pd.read_csv("data/water_features.csv", index_col="_id")
water_features.head(5)
```

```python
count = water_features['neighborhood'].value_counts().rename_axis('neighborhood').reset_index(name='count')
print(count.head(15))
sorted = count.sort_values(by='count', ascending=False)
file_path = 'final_metrics/sorted_water_counts_by_neighborhood.csv'
sorted.to_csv(file_path, index=False)
```

```python
import matplotlib.pyplot as plt
ft = pd.DataFrame(count).head(15)
ft.plot(x='neighborhood', y='count', kind='bar')
plt.title('Water Features in Pittsburgh Neighborhoods')
plt.xlabel('Neighborhoods')
plt.ylabel('Features')
plt.show()
```

South Squirrel Hill is the neighborhood with the most water features. This isn't very surprising, since Squirrel hill is one of the more popular neighborhoods in Pittsburgh. Highland Park isn't too far behind, with 12, but then it drops off with a large number of neighborhoods having between 6-8 water features.
